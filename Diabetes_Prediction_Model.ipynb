```python
# import libraries

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler # to standardize the data (scale them to be within the same range, usually betwen 0 and 1)
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# Data Collection and Analysis

```python
# I'm using PIMA Diabetes Dataset (it contains info for only females)

data = pd.read_csv('/content/diabetes.csv')
```

0 --> Non-diabetic
1 --> Diabetic

```python
data.groupby('Outcome').mean()
```

```python
# separate features and labels

X = data.drop(columns='Outcome', axis=1)
y = data['Outcome']
```

```python
print(X)
print('')
print(y.shape)
```

# Data Preprocessing

```python
# Data Standardization

scaler = StandardScaler()

scaler.fit(X) # let's just say this step is used to get a 'feel' of the data and understand the ranges

standardized_data = scaler.transform(X) # this step is what actually scales them

# you can also use scaler.fit_transform(X) to combine the steps
```

```python
print(standardized_data)
```

```python
X = standardized_data # saving the new standardized data back to X

print(X)
print(y) # we don't need to standardize y because it falls within the range of 0 and 1 (it's actually exactly 0 and 1 lol)
```

# Split Data into Train and Test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)

# stratify=y to make sure there are about the same number of y across test and train (similar proportions)

print(f'X_train: {X_train.shape}')
print(f'X_test: {X_test.shape}')
print(f'y_train: {y_train.shape}')
print(f'y_test: {y_test.shape}')
```

```python
print(f'X: {X.shape}')
```

# Model Training

```python
classifier = svm.SVC(kernel='linear') # kernel='linear' means we're going to use a linear model since we're going to use a linear line to sparate the data

# training the Support Vector Machine Classfier

classifier.fit(X_train, y_train)
```

# Model Evaluation

```python
# accuracy score on training data

X_train_pred = classifier.predict(X_train)

training_data_acc = accuracy_score(X_train_pred, y_train)
```

```python
print(f'Training data accuracy: {training_data_acc}')
```

```python
# accuracy score on testing data

X_test_pred = classifier.predict(X_test)

testing_data_acc = accuracy_score(X_test_pred, y_test)
```

```python
print(f'Testing data accuracy: {testing_data_acc}')
```

Looks like there's some overfitting going on (since the model performed better on the training than on the testing)

# Usage

```python
input_data = (6,148,72,35,0,33.6,0.627,50)

# we'll convert the input_data to a numpy array for faster and better processing
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array because we're predicting on one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Define the feature names (same as used during training)
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
           'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Create a DataFrame with the input data and matching feature names
input_df = pd.DataFrame(input_data_reshaped, columns=columns)

# we can't give the model that raw input_data because we standardized the training data when training the model. so we'll have to do the same here
std_data = scaler.transform(input_df) # we don't need to call fit() because the model has already gotten a 'feel' for the ranges during the scaling earlier

print(std_data)

print('')

prediction = classifier.predict(std_data)

print(prediction)

print('')

if prediction[0] == 1:
  print('Patient is Diabetic')
else:
  print('Patient is Non-diabetic')
```
